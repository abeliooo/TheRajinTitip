```python
import pandas as pd
import numpy as np
import gzip
import json
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = '/content/drive/MyDrive/TheRajinTItip/meta_Electronics.jsonl.gz'
limit = 500
```

```python
data_list = []
```

```python
with gzip.open(file_path, 'rt', encoding='utf-8') as f:
    for line in f:
      if len(data_list) >= limit:
        break
      try:
        data = json.loads(line)
        data_list.append(data)
      except json.JSONDecodeError:
        pass
```

```python
df_sample = pd.DataFrame(data_list)
df_sample.info()
```

```python
display(df_sample.head())
```

```python
df_sample = df_sample.dropna(subset=['price', 'images', 'details', 'categories'])
condition1 = df_sample['features'].apply(lambda x: x != [])
condition2 = df_sample['description'].apply(lambda x: x != [])

df_sample = df_sample[condition1 & condition2]
```

```python
df_product = df_sample.copy()
```

```python
column = ['title', 'features', 'description', 'price', 'images', 'categories', 'details', 'parent_asin']
```

```python
is_column = [col for col in column if col in df_product.columns]
df_product = df_product[is_column]
```

```python
display(df_product.head())
```

```python
df_product.info()
```

```python
def extract_image(image_list):
  if not isinstance(image_list, list) or not image_list:
    return None
  try:
    first_image = image_list[0]
    return first_image.get('hiRes', first_image.get('large'))
  except (KeyError, TypeError):
    return None
  return image_list[0]

df_product['image_url'] = df_product['images'].apply(extract_image)
```

```python
def extract_merk(merk_list):
  if not isinstance(merk_list, dict):
    return None
  try:
    return merk_list.get('Brand')
  except Exception:
    return None

df_product['merk'] = df_product['details'].apply(extract_merk)
```

```python
def unpack_description(desc):
    if isinstance(desc, list) and desc:
        return desc[0]
    return desc

df_product['description'] = df_product['description'].apply(unpack_description)
```

```python
def merge(row):
  current_description = row['description']
  if isinstance(row['features'], list) and row['features']:
        fitur_str = "\n".join([f"- {item}" for item in row['features']])
        return f"{current_description}\n\n{fitur_str}"

  else:
    return current_description

df_product['description'] = df_product.apply(merge, axis=1)
```

```python
df_product['name'] = df_product.apply(
    lambda row: f"{row['merk']} {row['title']}" if row['merk'] != 'Unknown' else row['title'],
    axis=1
)
```

```python
df_product = df_product.drop(columns=['images', 'details'])
df_product = df_product.rename(columns={'parent_asin':'id', 'price':'startingPrice', 'image_url':'image'})
```

```python
df_product.info()
```

```python
df_product = df_product.dropna(subset=['merk'])
```

```python
column = ['id', 'name', 'image', 'categories', 'description', 'startingPrice']
```

```python
df_product = df_product[column]
```

```python
df_product.info()
```

```python
df_product['startingPrice'] = df_product['startingPrice'] * 16500
# Price in Oct 20
display(df_product.head(50))
```

```python
df_product['currentPrice'] = df_product['startingPrice']
df_product['status'] = 'active'
df_product['isSold'] = False
```

```python
time_now = pd.to_datetime('now')
random = len(df_product)
random_days = np.random.randint(1, 8, size=random)
random_hours = np.random.randint(0, 24, size=random)
random_minutes = np.random.randint(0, 60, size=random)

random_days = pd.to_timedelta(random, unit='D')
random_hours = pd.to_timedelta(random_hours, unit='h')
random_minutes = pd.to_timedelta(random_minutes, unit='min')

df_product['auctionEndDate'] = time_now + random_days + random_hours + random_minutes
```

```python
df_product = df_product[column]
df_product.info()
```
