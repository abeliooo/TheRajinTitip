```python
import pandas as pd
import numpy as np
from faker import Faker
import random
```

```python
fake = Faker('id_ID')
```

```python
total = 100
data_user = []
```

```python
for _ in range(total):
  user = {
        'username': fake.user_name(),
        'fullName': random.choice([fake.name_male(), fake.name_female()]),
        'email': fake.email(),
        'phoneNumber': fake.numerify(text='08##########'),
        'address': fake.address(),
        'password': 'password123',
        'accountNumber': fake.numerify(text='##########'),
        'isAdmin': False
    }
  data_user.append(user)
```

```python
df_users = pd.DataFrame(data_user)
```

```python
display(df_users.head(10))
```

```python
name = 'therajintitip_user.csv'

df_users.to_csv(name, index=False)
```
